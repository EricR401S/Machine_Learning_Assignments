### **1b) Preprocessing the Data**

##### The strategies are the following : imputation, standardization and specific uses of PCA. For PCA, I am planning to treat it as a hyperparameter and tune it. Dimensionality reduction always implies a loss in information and a potential gain or loss in predictive power, which justifies the use of tuning. Then, I will prepare my train and validation split.

##### As for the in-depth details of the preprocessing, the very necessary one-hot encoding was already provided. From there, the missing values were imputed by using the median strategy via SimpleImputer, and the data was standardized with StandardScaler due to the difference in scales, such as years (4 digit values) vs. number of children (one digit value) vs. arrival time (3 digit values at most). I implement it through a pipeline and preprocess the data once. As for PCA, I will tune it on a case by case basis, but it does get piped when I run a Neural Network implementation, due to the nuances of the random search module. Typically, with PCA, I have to accept an inherent risk of accidentally losing useful features when accepting a certain threshold of variance explained by the remaining features. On the other hand, the opposite is true; one can have enough variance explained with minimal dimensionality reduction, which goes against the purpose of PCA in this context. That's why I decide to tune it as a hyperparameter, since I don't know what the right choice is. 

```python
# preprocessing on the data
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline

# define the pipeline
pipeline = Pipeline(
    [("imputer", SimpleImputer(strategy="median")), ("scaler", StandardScaler())]
)
pipeline.fit(X_train_ohe)
X_train_ohe_scaled = pipeline.transform(X_train_ohe.copy())
X_test_ohe_scaled = pipeline.transform(X_test_ohe.copy())
```

```python
# preparing the train and validation sets
from sklearn.model_selection import train_test_split

X_train_hr, X_val, y_train_hr, y_val = train_test_split(
    X_train_ohe_scaled, y_train, test_size=0.2, random_state=42, stratify=y_train
)


# creating a representative subset of data

from sklearn.utils import resample

X_train_sample = resample(
    X_train_ohe_scaled.copy(),
    n_samples=5000,
    random_state=209,
    stratify=y_train.copy(),
)
y_train_sample = resample(
    y_train.copy(),
    n_samples=5000,
    random_state=209,
    stratify=y_train.copy(),
)

# splitting into a smaller training set and a validation set
X_train_sample, X_val_sample, y_train_sample, y_val_sample = train_test_split(
    X_train_sample,
    y_train_sample,
    test_size=0.2,
    random_state=42,
    stratify=y_train_sample,
)
```

### **1c) Select, train and compare models**

##### The strategy for this step is multi-layered. First, for each model, using a sample of the data, I will manually random search a space of parameters for a Logistic Regression model, a KNN model, a Neural Network, a Random Forest, and an XGboost model (a variant of gradient boosted decision trees). First, besides the regular tuning, tuning for the PCA transformation will be completed only for the non-decision tree models. My reasoning is that I simply won't know what number of features kept will be the best, so I am treating it as a hyperparamenter for PCA. As for the decision tree models, they have a robust dimensionality reduction as part of their algorithms, so I will defer the dimensionality reduction to them and will not use PCA for them. In terms of justification for PCA for the other models, PCA is a great tool for dimensionality reduction, especially when one can't gauge how important or unimportant a feature is in a problem space. ##### My guiding metric for the winning hyperparameter is the AUC. AUC is special because it tracks the True Positive Rate (TPR) and the False Positive Rate (FPR). It is a metric that tries to maximize the best TPR while keeping the FPR low. This is especially true for a simple reason. The positive class is the minority class, which I will show in the following code block. Maximizing the TPR is a priority given that the client is highly likely to want to accurately know which person will cancel the booking. A low FPR will implicitely indicate that the model is trying to keep a low misclassification of negatives as positives. In other words, this metric serves to inform accuracy in ways beyond a simple accuracy score. If my model has excellent accuracy, but a horrible capacity to predict the positive class, then this reliance may become an issue (and I'm also aware that simply relying on AUC has its issues as well). No one metric solves every problem, but I chose it as the one to best inform my model selection.

##### The following function is essential to produce the best models after tuning. It trains the tuned model on the training set and runs tests on the validations set (auc and accuracy). It also records the training and prediction time for the final models. This function fulfills the latter half of the prompt for part c.

```python
# Function to avoid repetitive code and human error
def final_model(best_model, name, final_pca, X_train_hr = X_train_hr, y_train_hr = y_train_hr, X_val = X_val, y_val =  y_val, pipe = False):
    """Fit the best model on the training set and evaluate it on the validation set
    
    It takes the best model and the best PCA scaler and fits the model on the training set
    and evaluates it on the validation set. 
    This only works for my non XGBoost models.
    It will return two dictionaries. 
    One records the training and testing times and 
    the other will return the final roc and pr curve arrays,
    as well as the final average precision score, the final auc score, and the final accuracy score.
    """

    from sklearn.metrics import roc_auc_score, roc_curve
    from sklearn.metrics import precision_recall_curve, average_precision_score
    import time

    if pipe is False:
        # fit the best PCA scaler
        final_pca.fit(X_train_hr)
        X_train_hr_pca = final_pca.transform(X_train_hr)
        X_val_pca = final_pca.transform(X_val)

    else:
        # to accomodate the neural network pipeline
        X_train_hr_pca = X_train_hr
        X_val_pca = X_val


    # fit the model
    train_start = time.time()
    best_model.fit(X_train_hr_pca, y_train_hr)
    train_end = time.time()
    print("The training time for the final " + name+ " model is {:.2f} seconds.".format(train_end - train_start))

    # predict the probabilities
    predict_start = time.time()
    best_model_proba = best_model.predict_proba(X_val_pca)[:, 1]
    predict_end = time.time()
    print("The prediction time for the final " + name + " model is {:.2f} seconds.".format(predict_end - predict_start))

    # Add the time consumption to the dictionary
    time_dictionary = {'train_time' : train_end - train_start, 'predict_time' : predict_end - predict_start}

    # calculate the auc and accuracy and ap
    best_model_auc = roc_auc_score(y_val, best_model_proba)
    best_model_accuracy = best_model.score(X_val_pca, y_val)
    best_model_ap = average_precision_score(y_val, best_model_proba)

    # roc curve for the final logit model
    fpr, tpr, roc_thresholds = roc_curve(y_val, best_model_proba)
    # pr curve for the final logit model
    precision, recall, pr_thresholds = precision_recall_curve(y_val, best_model_proba)

    # add the results to the dictionary
    final_model_dictionary = {'fpr' : fpr, 'tpr' : tpr, 'roc_thresholds' : roc_thresholds, 
                                'pr_thresholds' : pr_thresholds, 'auc' : best_model_auc, 'accuracy' : best_model_accuracy,
                                'ap' : best_model_ap, 'precision' : precision, 'recall' : recall}

    return final_model_dictionary, time_dictionary


```

```python
# This PCA number of components list will be used often
# as a search space for this hyperparameter
PCA_list = np.linspace(0.1, 1, 10)
PCA_list = PCA_list[:-1]
PCA_list = np.append(PCA_list, 0.95)

# Global variable for the number of search iterations
n_iter = 100

# this will house the roc curves and pr curves for the final models
# as well as the training and prediction times
final_model_dictionary = {}
time_dictionary = {}
```

#### Hyperparameter Tuning : Logistic Regression

>Note : According to the documentation, the solver in logistic regression learns the learning rate, so I will focus on tuning the number of components for PCA and the regularization strength C for the logistic regression classifier. 

```python
# PCA tuning for logistic regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
from sklearn.decomposition import PCA

# hyperparameter tuning the logistic regression model
# According to documentation, the solver learns the learning rate
# So the tuning will focus on the reg
C_list = np.logspace(-4, 4, 20)


# create a list to store the results
logit_auc = {}
logit_accuracy = {}
logit_params = {}
logit_pca = {}


# random search for the best hyperparameters
for i in range(n_iter):

    # for reproducibility
    np.random.seed(i)

    run_C = np.random.choice(C_list)
    run_PCA = np.random.choice(PCA_list)

    # fit the PCA scaler
    pca_scaler_for_logit = PCA(n_components=run_PCA, random_state=42)
    pca_scaler_for_logit.fit(X_train_sample)
    X_train_sample_pca = pca_scaler_for_logit.transform(X_train_sample)
    X_val_sample_pca = pca_scaler_for_logit.transform(X_val_sample)

    # logistic regression classifier
    # newton cholesky is better for datasets with tons of one hot encoded features
    logit = LogisticRegression(solver = 'newton-cholesky', penalty = 'l2', C = run_C, random_state=42, n_jobs=-1)

    # fit the pipeline
    logit.fit(X_train_sample_pca, y_train_sample)

    auc = roc_auc_score(y_val_sample, logit.predict_proba(X_val_sample_pca)[:, 1])
    score = logit.score(X_val_sample_pca, y_val_sample)

    # record the results
    logit_auc[i] = auc
    logit_accuracy[i] = score
    logit_params[i] = {'C' : run_C}
    logit_pca[i] = {'n_components' : run_PCA}

# best AUC key

best_logit_auc = max(logit_auc, key=logit_auc.get)
best_logit_accuracy = max(logit_accuracy, key=logit_accuracy.get)
top_5_logit_auc_keys = sorted(logit_auc, key=logit_auc.get, reverse=True)[:5]
best_logit_key = top_5_logit_auc_keys[0]

# print the AUC and accuracy scores
print("The best AUC score is {:.4f}.".format(logit_auc[best_logit_auc]))
print("The best accuracy score on the validation set  is {:.4f}.".format(logit_accuracy[best_logit_accuracy]))

# print the best parameters chosen by the highest AUC score
print("The best hyperparameters are {}.".format(logit_params[best_logit_auc]))

final_logit_model = LogisticRegression(solver = 'newton-cholesky', penalty = 'l2', C = logit_params[best_logit_auc]['C'], random_state=42, n_jobs=-1)
final_logit_pca = PCA(n_components=logit_pca[best_logit_key]['n_components'], random_state=42)
```

```python
# Training and validating with the best logit model

final_model_dictionary['logit'], time_dictionary['logit'] = final_model(final_logit_model, 'logit', final_logit_pca, X_train_hr, y_train_hr, X_val, y_val)
```

#### Hyperparameter Tuning : K-Nearest Neighbors

##### The main hyperparameter to tune for KNN is the number of neighbors, so that will be the focus in the next cell block. 

```python
# PCA tuning for KNN
from sklearn.neighbors import KNeighborsClassifier

from sklearn.decomposition import PCA

# hyperparameter tuning the KNN
# the focus is on the number of neighbors
# and the PCA components

k_list = np.linspace(1, X_train_sample.shape[0], 100, dtype = int)


# create a list to store the results
knn_auc = {}
knn_accuracy = {}
knn_params = {}
knn_pca = {}


# random search for the best hyperparameters
for i in range(n_iter):

    # for reproducibility
    np.random.seed(i)

    run_k = np.random.choice(k_list)
    run_PCA = np.random.choice(PCA_list)

    # fit the PCA scaler
    pca_scaler_for_knn = PCA(n_components=run_PCA, random_state=42)
    pca_scaler_for_knn.fit(X_train_sample)
    X_train_sample_pca = pca_scaler_for_knn.transform(X_train_sample)
    X_val_sample_pca = pca_scaler_for_knn.transform(X_val_sample)

    # logistic regression classifier
    # newton cholesky is better for datasets with tons of one hot encoded features
    knn = KNeighborsClassifier(n_neighbors=run_k, n_jobs=-1)

    # fit the pipeline
    knn.fit(X_train_sample_pca, y_train_sample)

    auc = roc_auc_score(y_val_sample, knn.predict_proba(X_val_sample_pca)[:, 1])
    score = knn.score(X_val_sample_pca, y_val_sample)

    # record the results
    knn_auc[i] = auc
    knn_accuracy[i] = score
    knn_params[i] = {'k' : run_k}
    knn_pca[i] = {'n_components' : run_PCA}

# best AUC key

best_knn_auc = max(knn_auc, key=knn_auc.get)
best_knn_accuracy = max(knn_accuracy, key=knn_accuracy.get)
top_5_knn_auc_keys = sorted(knn_auc, key=knn_auc.get, reverse=True)[:5]
best_knn_key = top_5_knn_auc_keys[0]

# print the AUC and accuracy scores
print("The best AUC score is {:.4f}.".format(knn_auc[best_knn_auc]))
print("The best accuracy score on the validation set  is {:.4f}.".format(knn_accuracy[best_knn_accuracy]))

# print the best parameters chosen by the highest AUC score
print("The best hyperparameters are {}.".format(knn_params[best_knn_auc]))

final_knn_model = KNeighborsClassifier(n_neighbors=knn_params[best_knn_key]['k'], n_jobs=-1)
final_knn_pca = PCA(n_components=knn_pca[best_knn_key]['n_components'], random_state=42)
```

##### Note : It is important to convert or extrapolate the ratio of the k from the sample for the validation run and the final model submission.

#### Hyperparameter Tuning : Neural Networks

##### The chosen hyperparameters to tune for Neural Networt (MLP Classifier) is the learning rate, the alpha or strength of the regularization term, the batch size, the nodes, and number of layers. Tuning it will require a long time, and I will make use of the random search utility from sklearn, as the MLP classifier unfortunately lacks the n_jobs argument.

```python

nodes = [1, 2, 3, 4, 5, 10, 15, 25, 30]
layers_nums = [1, 2, 3, 4]

def node_layer_config(nodes_per_layer, layers_num):
    """Build the nodes-layers configuration to serve as a hyperparameter for the neural network model"""
    list_config = (np.ones((layers_num)) * nodes_per_layer).tolist()
    list_config = [int(i) for i in list_config]
    tuple_config = tuple(list_config)
    return tuple_config

nodes_layer_grid = []

for nodes_per_layer in nodes:
    for layers_num in layers_nums:
        nodes_layer_grid.append(node_layer_config(nodes_per_layer, layers_num))
```

```python
# new pipeline for the neural network
from sklearn.neural_network import MLPClassifier

pipe_nn = Pipeline(
    [
        ("pca", PCA(random_state=42)),
        (
            "nn",
            MLPClassifier(
                solver="adam",  # For large datasets, ‘adam’ is the preferred solver.
                tol=1e-5,
                early_stopping=False,
                activation="relu",
                n_iter_no_change=1000,
                max_iter=500,
                random_state=42,
            ),
        ),
    ]
)
```

```python
# For RandomSeachCV, combine training and validation set samples 
# and specify which portion is training and which is validation
from sklearn.model_selection import PredefinedSplit
sample_X_train_plus_val = np.concatenate((X_train_sample , X_val_sample), axis=0)
sample_y_train_plus_val = np.concatenate((y_train_sample, y_val_sample), axis=0)

# Create a predefined train/test split for RandomSearchCV 
validation_fold = np.concatenate((-1 * np.ones(len(y_train_sample)), np.zeros(len(y_val_sample))))
train_val_split = PredefinedSplit(validation_fold)
```

```python
import numpy as np

from time import time
import scipy.stats as stats
from sklearn.utils.fixes import loguniform

from sklearn.model_selection import RandomizedSearchCV


# Utility function to report best scores
def report(results, n_top=3):
    for i in range(1, n_top + 1):
        candidates = np.flatnonzero(results["rank_test_score"] == i)
        for candidate in candidates:
            print("Model with rank: {0}".format(i))
            print(
                "Mean validation score: {0:.3f} (std: {1:.3f})".format(
                    results["mean_test_score"][candidate],
                    results["std_test_score"][candidate],
                )
            )
            print("Parameters: {0}".format(results["params"][candidate]))
            print("")


# specify parameters and distributions to sample from
param_dist = {
    "nn__learning_rate_init" : loguniform(1e-5, 1e0),
    "nn__batch_size": [1, 3, 5, 10, 20, 50, 100, 250, 500],
    "nn__alpha": loguniform(1e-8, 1e2),
    "nn__hidden_layer_sizes": nodes_layer_grid,
    "pca__n_components": PCA_list,
}

# run randomized search
n_iter_search = n_iter
random_search = RandomizedSearchCV(
    pipe_nn,
    param_distributions=param_dist,
    n_iter=n_iter_search,
    n_jobs=-1,
    cv=train_val_split,
    scoring="roc_auc",
)

start = time()
random_search.fit(sample_X_train_plus_val, sample_y_train_plus_val)
print(
    "RandomizedSearchCV took %.2f seconds for %d candidates parameter settings."
    % ((time() - start), n_iter_search)
)
report(random_search.cv_results_)
```

```python
################################
# Code to plot clusters
################################
def plot_cluster(ax, data, cluster_assignments):
    '''Plot two-dimensional data clusters
    
    Parameters
    ----------
    ax : matplotlib axis
        Axis to plot on
    data : list or numpy array of size [N x 2] 
        Clustered data
    cluster_assignments : list or numpy array [N]
        Cluster assignments for each point in data

    '''
    clusters = np.unique(cluster_assignments)
    n_clusters = len(clusters)
    for ca in clusters:
        kwargs = {}
        if ca == -1:
            # if samples are not assigned to a cluster (have a cluster assignment of -1, color them gray)
            kwargs = {'color':'gray'}
            n_clusters = n_clusters - 1
        ax.scatter(data[cluster_assignments==ca, 0], data[cluster_assignments==ca, 1],s=5,alpha=0.5, **kwargs)
        ax.set_xlabel('feature 1')
        ax.set_ylabel('feature 2')
        ax.set_title(f'No. Clusters = {n_clusters}')
        ax.axis('equal')
```

##### Now to produce an elbow curve of the K-Means clustering.

```python
from sklearn.cluster import KMeans

# a list of k-values and inertial values

k_values = [i for i in range(1, 11)]

name_dn = {"X0": X0, "X1": X1, "X2": X2, "X3": X3, "X4": X4}
inertia_dn = {}


for ds in name_dn:

    inertia_values = []

    for k in k_values:

        dataset = name_dn[ds]
        
        kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)

        kmeans.fit(dataset)

        # append the inertia to the list of inertias

        inertia_values.append(kmeans.inertia_)

    inertia_dn[ds] = inertia_values
```

```python
# plot the elbow curves

fig, ax = plt.subplots(1, 5, figsize=(28, 5))

for i, ds in enumerate(name_dn):
    
        ax[i].plot(k_values, inertia_dn[ds], marker='o')
        ax[i].set_xlabel('Number of clusters')
        ax[i].set_ylabel('Inertia')
        ax[i].set_title(f'Elbow Curve for {ds}')
```
